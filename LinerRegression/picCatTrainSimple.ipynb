```python
from LinerRegression.picCatTrainSet import load_data_fashion_mnist
import warnings

warnings.filterwarnings("ignore")
import paddle
from paddle import nn

batch_size = 256
train_iter, test_iter = load_data_fashion_mnist(batch_size)
```

```python
net = nn.Sequential(nn.Flatten(), nn.Linear(784, 10))

def init_weights(m):
    if type(m) == nn.Linear:
        nn.initializer.Normal(m.weight, std=0.01)

net.apply(init_weights);
```

```python
loss = nn.CrossEntropyLoss(reduction='none')
```

```python
trainer = paddle.optimizer.SGD(learning_rate=0.2, parameters=net.parameters())
```

```python
from LinerRegression.picCatTrainFromZeroUtils import train_ch3

num_epochs = 10
train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```
